```python
def create_features_forecasting(df_incomes):
    """
    Создает признаки для задачи прогнозирования на основе временных рядов внесений.
    
    df_incomes: DataFrame, индекс = TID, столбцы = даты
    
    Возвращает: датафрейм в формате long (TID, date, features)
    """
    # Переводим банкоматы в строки
    df = df_incomes.copy().T
    df.index = pd.to_datetime(df.index)
    
    # Сохраняем исходные данные
    df_long = df.stack().reset_index()
    df_long.columns = ['date', 'TID', 'income']

    # Добавляем признаки времени
    df_long['day_of_week'] = df_long['date'].dt.dayofweek  # день недели
    df_long['is_weekend'] = df_long['day_of_week'].isin([5, 6]).astype(int)  # выходной или нет

    # Сортируем для корректных лагов
    df_long = df_long.sort_values(['TID', 'date'])

    # Добавляем лаги
    for lag in [1, 2, 3, 7, 14]:
        df_long[f'lag_{lag}'] = df_long.groupby('TID')['income'].shift(lag)
    
    # Добавляем скользящие средние
    for window in [3, 7, 14]:
        df_long[f'rolling_mean_{window}'] = df_long.groupby('TID')['income'].shift(1).rolling(window).mean().reset_index(level=0, drop=True)
    
    # Можно также добавить скользящие стандартные отклонения (по желанию)

    # Убираем пропуски (где не хватает данных для лагов/средних)
    df_long = df_long.dropna()

    return df_long
```

```python
def create_features_clustering(df_incomes):
    """
    Создает агрегированные признаки для кластеризации банкоматов.
    
    df_incomes: DataFrame, индекс = TID, столбцы = даты
    
    Возвращает: датафрейм с признаками по каждому банкомату
    """
    features = pd.DataFrame(index=df_incomes.index)

    features['mean_income'] = df_incomes.mean(axis=1)
    features['std_income'] = df_incomes.std(axis=1)
    features['max_income'] = df_incomes.max(axis=1)
    features['min_income'] = df_incomes.min(axis=1)
    features['zero_days_ratio'] = (df_incomes == 0).sum(axis=1) / df_incomes.shape[1]
    features['days_over_100k'] = (df_incomes > 100000).sum(axis=1)

    # Добавить сюда геолокацию, если надо
    # features = features.merge(df_tids, on='TID', how='left')

    return features
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 2. Загружаем данные (у тебя уже есть df_locations)
df_locations = pd.read_excel('data/gpb_adm_data.xlsx', sheet_name=0)

# 3. Выбираем только координаты
X = df_locations[['longitude', 'latitude']]

# 4. Обучаем KMeans
n_clusters = 10  # сколько хочешь кластеров
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df_locations['cluster'] = kmeans.fit_predict(X)

# 5. Визуализация
plt.figure(figsize=(10, 8))
plt.scatter(df_locations['longitude'], df_locations['latitude'], c=df_locations['cluster'], cmap='tab10', s=30)
plt.title('Кластеры банкоматов')
plt.xlabel('Долгота')
plt.ylabel('Широта')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time

# Функция для парсинга по TID
def get_schedule(tid):
    url = f"https://www.gazprombank.ru/atms/{tid}/"
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        div = soup.find('div', class_='card__schedule-207')
        if div:
            return div.get_text(strip=True)
        else:
            return "Нет данных"
    except Exception as e:
        return f"Ошибка: {e}"

# Создадим новую колонку с расписанием
df_tids['schedule'] = df_tids.head(20)['TID'].apply(get_schedule)


print(df_tids)
```
